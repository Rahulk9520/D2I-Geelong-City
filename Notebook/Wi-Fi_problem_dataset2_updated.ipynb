# Data Cleaning


```python
#Filling NaN values in "Device name" column with most frequent column values
# defining the imputer
imputer = SimpleImputer(missing_values=nan, strategy='most_frequent')
 
# transform the dataset
df['Device name'] = imputer.fit_transform(df['Device name'].values.reshape(-1,1))[:,0]
```

```python
#Filling NaN values in "Device name" column with most frequent column values
# defining the imputer
imputer = SimpleImputer(missing_values=nan, strategy='most_frequent')
 
# transform the dataset
df['devicetype'] = imputer.fit_transform(df['devicetype'].values.reshape(-1,1))[:,0]
```

```python
#Dropping the "location" column with NaN values
df = df.dropna(subset = ["location"])
```
