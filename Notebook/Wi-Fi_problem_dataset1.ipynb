# Data Cleaning


```python
#Dropping the "dev_id" column with NaN values
df = df.dropna(subset = ["dev_id"])
```

```python
#Dropping the "name" column with NaN values
df = df.dropna(subset = ["name"])
```

```python
#Dropping the "date" column with NaN values
df = df.dropna(subset = ["date"])
```

```python
#Filling NaN values in "total" column with mean column values
df['total'].fillna(value=df['total'].mean(), inplace=True)
```

```python
#Filling NaN values in "left" column with mean column values
df['left'].fillna(value=df['left'].mean(), inplace=True)
```

```python
#Filling NaN values in "left_avg" column with mean column values
df['left_avg'].fillna(value=df['left_avg'].mean(), inplace=True)
```

```python
#Checking the null values again
new_null_val = pd.DataFrame({'Null Values' : df.isna().sum(), 'Percentage Null Values' : (df.isna().sum()) / len(df) * (100)})
new_null_val
```
