##  Importing important packages and libraries ##

```python
import numpy as np
import os
import math
from numpy import nan
from numpy import mean
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter(action='ignore')
```

### Importing cleaned Wi-Fi datset-1 (from 2018 to Nov 2020)

```python
# importing the cleaned_wifi_data dataset
df = pd.read_csv("D:/Deakin University (Applied AI (Professional)/SIT764- Team Project (A)/Datasets & Notebooks/Updated Dataset/cleaned_wifi_data.csv")
df = df.drop('Unnamed: 0', axis=1)
df.head(5)
```

### Importing cleaned Wi-Fi datset-2 (from Nov 2020 till 7th Dec 2021)


```python
# importing the cleaned_wifi_data1 dataset
df1 = pd.read_csv("D:/Deakin University (Applied AI (Professional)/SIT764- Team Project (A)/Datasets & Notebooks/Updated Dataset/regularupdated_cleaned_wifi_data.csv")
df1 = df1.drop('Unnamed: 0', axis=1)
df1.head(5)
```

```python
# displaying the columns of regularupdated_cleaned_wifi_data dataset 
pd.set_option("display.max_columns", None)
df1.head(1)
```

```python
# removing the columns form regularupdated_cleaned_wifi_data which are not present in cleaned_wifi_data dataset
# storing the results in new_df1
new_df1 = df1.drop(['devicetype','avg_total', 'new','hist_1','hist_2',
                    'hist_3','hist_4','hist_5','hist_6','hist_7','hist_8'], axis = 1)
new_df1.head(5)
```

```python
# renaming the columns of regularupdated_cleaned_wifi_data dataset as per cleaned_wifi_data dataset
new_df1.rename(columns = {'max_total':'total','left_avg_mins (dwell time)':'left_avg','dayOfWeek':'dayofweek',
                          'Device id':'dev_id','Device name':'name','time':'date','daynumber':'day'},inplace = True)
```

```python
# removing the extra columns in cleaned_wifi_data dataset which are not present in regularupdated_cleaned_wifi_data dataset
new_df = df.drop(['updatetime','hour'], axis = 1)
new_df.head(5)
```

```python
# checking wheather the columns are matching for both dataframes

print("Updated attributes of cleaned_wifi_data dataset")
for col in new_df.columns:
    print(col)
    
print("\n Updated attributes of regularupdated_cleaned_wifi_data dataset")
for col in new_df1.columns:
    print(col)
```

```python
#checking the length of the Cleaned Wifi datasets
print(len(new_df))
print(len(new_df1))
```

### Merging operation of both cleaned datsets

```python
append_df = new_df.append(new_df1, ignore_index = True)
append_df
```

```python
# checking for null values 

null_val = pd.DataFrame({'Null Values' : append_df.isna().sum(), 'Percentage Null Values' : (append_df.isna().sum()) / len(append_df) * (100)})
null_val
```

```python
#Checking the info of our Merged dataset 
append_df.info()
```

## Saving the cleaned dataset to a CSV file

```python
#Saving the Cleaned wifi_data to a CSV file
append_df.to_csv('D:/Deakin University (Applied AI (Professional)/SIT764- Team Project (A)/Datasets & Notebooks/Updated Dataset/Merged_cleaned_wifi_data.csv')
print('Merged cleaned Wi-Fi dataset has been saved sccessfully')
```
